Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its
application.

Ans:
    
    Min-Max scaling is a data normalization technique used in data preprocessing to scale numeric features within a 
    specific range, typically between 0 and 1. The purpose of Min-Max scaling is to transform the original data so that
    it falls within a specific interval, making it easier to compare and analyze features with different units and scales.

    The Min-Max scaling formula for a feature X is given by:
        Xscaled = X - min(X)/max(X) - min(X)
        
    where:
    X scaled is the scaled value of the feature 
    min(X) is the minimum value of X
    max(X) is the maximum value of X

```python
# Example
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Sample data (replace this with your own dataset)
data = np.array([[2.0, 5.0],
                 [10.0, 20.0],
                 [5.0, 15.0]])

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the data using Min-Max scaling
scaled_data = scaler.fit_transform(data)

# Display the original and scaled data
print("Original Data:")
print(data)
print("\nScaled Data:")
print(scaled_data)
```

Q2. What is the Unit Vector technique in feature scaling, and how does it differ from Min-Max scaling?
Provide an example to illustrate its application.

Ans:
    
    The unit vector technique, also known as vector normalization or unit vector scaling, is a feature scaling method that
    involves scaling the values of a feature to create a unit vector. In other words, it scales the feature in such a way 
    that the resulting vector has a length of 1 while preserving its direction. The formula for unit vector scaling is as
    follows:
    Xscaled=X/∥X∥
    
    where:
    Xscaled is the scaled value of the feature 
    X is the original value of the feature,
    ∥X∥ is the Euclidean norm (length) of the vector 
    The unit vector technique is particularly useful when the direction of the data points is important, but the scale is
    not critical.

    Here's an example in Python illustrating unit vector scaling:


```python
# example
import numpy as np
from sklearn.preprocessing import Normalizer

# Sample data (replace this with your own dataset)
data = np.array([[2.0, 5.0],
                 [10.0, 20.0],
                 [5.0, 15.0]])

# Create a Normalizer object
normalizer = Normalizer()

# Fit and transform the data using unit vector scaling
unit_vector_scaled_data = normalizer.fit_transform(data)

# Display the original and scaled data
print("Original Data:")
print(data)
print("\nUnit Vector Scaled Data:")
print(unit_vector_scaled_data)
```

Q3. What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an
example to illustrate its application.

Ans:
    
    Principal Component Analysis (PCA) is a dimensionality reduction technique commonly used in machine learning and 
    statistics. The goal of PCA is to transform the original features of a dataset into a new set of linearly uncorrelated 
    variables, called principal components. These components are ordered by the amount of variance they explain in the 
    data, with the first principal component capturing the maximum variance and each subsequent component capturing less.

    The steps involved in PCA are as follows:

    Standardize the Data: If the features of the dataset are measured on different scales, it's common practice to 
        standardize them to have zero mean and unit variance.

    Compute the Covariance Matrix: Calculate the covariance matrix for the standardized data.

    Compute Eigenvectors and Eigenvalues: Find the eigenvectors and eigenvalues of the covariance matrix.

```python
# example
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load the Iris dataset (replace this with your own dataset)
iris = load_iris()
X = iris.data
y = iris.target

# Standardize the data
X_standardized = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Apply PCA with 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_standardized)

# Display the original and reduced-dimensional data
print("Original Data Shape:", X.shape)
print("Reduced Data Shape:", X_pca.shape)
```

Q4. What is the relationship between PCA and Feature Extraction, and how can PCA be used for Feature
Extraction? Provide an example to illustrate this concept.

Ans:
    
    Principal Component Analysis (PCA) can be used for feature extraction in the context of dimensionality reduction. 
    Feature extraction involves transforming the original features of a dataset into a new set of features, which are 
    often a reduced representation capturing the essential information. PCA achieves this by identifying the principal
    components, which are linear combinations of the original features.

    The relationship between PCA and feature extraction can be summarized as follows:

    Identifying Important Features: PCA identifies the directions (principal components) in which the data varies the most.
        These principal components are linear combinations of the original features, and they can be viewed as new, derived
        features.

    Reducing Dimensionality: By selecting a subset of the principal components that capture most of the variance in the data,
        PCA effectively reduces the dimensionality of the feature space. The new set of features (principal components) can
        be considered as a compressed representation of the original features.

```python
#Example
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load the digits dataset (replace this with your own dataset)
digits = load_digits()
X = digits.data
y = digits.target

# Apply PCA for feature extraction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualize the reduced-dimensional data
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.title('PCA for Feature Extraction')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(*scatter.legend_elements(), title='Digit Classes')
plt.show()
```

Q7. For a dataset containing the following values: [1, 5, 10, 15, 20], perform Min-Max scaling to transform the
values to a range of -1 to 1.

```python
import numpy as np

# Given dataset
data = np.array([1, 5, 10, 15, 20])

# Define the desired range
a = -1
b = 1

# Min-Max scaling
scaled_data = (data - np.min(data)) / (np.max(data) - np.min(data)) * (b - a) + a

print("Original Data:", data)
print("Min-Max Scaled Data (-1 to 1):", scaled_data)
```
